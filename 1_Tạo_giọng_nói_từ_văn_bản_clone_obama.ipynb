```python
# @title 10.TTS hàng loạt bằng giọng clone obama
from TTS.api import TTS
import glob
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Lấy danh sách tất cả các file âm thanh mẫu trong thư mục chỉ định
speaker_wavs = glob.glob('/content/drive/My Drive/project1/obamaclone/*.wav')

# Lấy danh sách tất cả các file văn bản trong thư mục chỉ định
text_files = glob.glob('/content/drive/My Drive/project1/*.txt')

# Khởi tạo mô hình TTS (sử dụng phiên bản mới nhất của mô hình xTTS)
tts = TTS(model_name="xtts", gpu=True)

# Duyệt qua từng file văn bản và chuyển đổi sang giọng nói
for idx, text_file in enumerate(text_files):
    with open(text_file, 'r') as file:
        text_to_convert = file.read()

    # Tạo giọng nói từ văn bản và lưu vào file
    output_path = f"/content/drive/My Drive/project1/output_{idx}.wav"
    tts.tts_to_file(text=text_to_convert, file_path=output_path, speaker_wav=speaker_wavs, language="en")
```
