```python
# @title 7. Chuyển đổi tất cả các file văn bản trong thư mục /coquitts trên Google Drive thành giọng nói và lưu các file âm thanh vào cùng thư mục
import os
import torch
from TTS.api import TTS

# Thiết lập thiết bị
device = "cuda" if torch.cuda.is_available() else "cpu"

# Khởi tạo mô hình TTS
tts = TTS(model_name="tts_models/en/ljspeech/tacotron2-DDC", progress_bar=False).to(device)

# Đường dẫn đến thư mục chứa file văn bản trên Google Drive
text_folder_path = '/content/drive/My Drive/coquitts'

# Duyệt qua tất cả các file văn bản trong thư mục
for file_name in os.listdir(text_folder_path):
    if file_name.endswith('.txt'):
        file_path = os.path.join(text_folder_path, file_name)
        with open(file_path, 'r') as file:
            text = file.read()

        # Tạo đường dẫn file âm thanh đầu ra
        output_audio_path = os.path.join(text_folder_path, file_name.replace('.txt', '.wav'))

        # Chuyển đổi văn bản thành giọng nói và lưu vào file
        tts.tts_to_file(text=text, file_path=output_audio_path)
```

```python
# @title 8.Chuyển đổi hàng loạt các file văn bản trong thư mục /coquitts/txt thành giọng nói sử dụng một mẫu âm thanh từ thư mục /coquitts/audioclone trên Google Drive
import os
import torch
from TTS.api import TTS

# Khởi tạo mô hình TTS
tts = TTS(model_name="tts_models/en/ljspeech/tacotron2-DDC", progress_bar=False).to(device)

# Đường dẫn đến thư mục chứa file văn bản và file âm thanh mẫu
text_folder_path = '/content/drive/My Drive/coquitts/txt'
audio_clone_path = '/content/drive/My Drive/coquitts/audioclone/speaker.wav'

# Duyệt qua tất cả các file văn bản trong thư mục
for file_name in os.listdir(text_folder_path):
    if file_name.endswith('.txt'):
        file_path = os.path.join(text_folder_path, file_name)
        with open(file_path, 'r') as file:
            text = file.read()

        # Tạo đường dẫn file âm thanh đầu ra
        output_audio_path = os.path.join(text_folder_path, file_name.replace('.txt', '_clone.wav'))

        # Chuyển đổi văn bản thành giọng nói và lưu vào file
        tts.tts_with_vc_to_file(text, audio_clone_path, file_path=output_audio_path)
```
